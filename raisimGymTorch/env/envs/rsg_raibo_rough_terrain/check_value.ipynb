```python
import pandas as pd

title = "box_fail"

control = pd.read_csv("control_{title}.csv".format(title=title))
raibot = pd.read_csv("raibot_{title}.csv".format(title=title))
```

```python
# print(subgoal_command_0.values)
import numpy as np
import matplotlib.pyplot as plt

subgoal_command_0 = control[" subgoal_command_0_0"]
subgoal_command_1 = control[" subgoal_command_1_0"]
subgoal_command_2 = control[" subgoal_command_2_0"]

value = 0
id = 0
while (True):
    value = subgoal_command_0[id]
    if(value != 0):
        break
    id += 1

subgoal_command_0 = subgoal_command_0[id:]
subgoal_command_1 = subgoal_command_1[id:]
subgoal_command_2 = subgoal_command_2[id:]

subgoal_command = np.array([subgoal_command_0.values, subgoal_command_1.values, subgoal_command_2.values])
```

```python
init_idx = 23900
end_idx = 51000

gc = np.zeros((19, len(raibot[' vicongenCoordinate_0_0'].values)))
for i in range(19):
    gc[i] = raibot[' vicongenCoordinate_{d}_0'.format(d=i)].values

object_tran = np.zeros((3, len(raibot[' viconObjectTran_0_0'].values)))
object_rot = np.zeros((9, len(raibot[' viconObjectTran_0_0'].values)))

for i in range(3):
    object_tran[i] = raibot[' viconObjectTran_{d}_0'.format(d=i)].values

for i in range(0,3):
    for j in range(0,3):
        object_rot[3*i + j] = raibot[' viconObjectRot_{i}_{j}'.format(i=i, j=j)]

gc = gc.transpose(1, 0)
object_tran = object_tran.transpose(1,0)
object_rot = object_rot.transpose(1,0)

gc = gc[init_idx:end_idx, :]
object_tran = object_tran[init_idx:end_idx, :]
object_rot = object_rot[init_idx:end_idx, :]

np.savetxt('/home/oem/gc_{title}'.format(title=title),gc)
np.savetxt('/home/oem/obj_tran_{title}'.format(title=title), object_tran)
np.savetxt('/home/oem/obj_rot_{title}'.format(title=title), object_rot)

x = np.linspace(0, gc.shape[0], gc.shape[0])

figure, ax = plt.subplots(figsize=(8,6))


line1, = ax.plot(x, gc[:, 0])
line2, = ax.plot(x, gc[:, 1])
line3, = ax.plot(x, gc[:, 2])

figure2, ax2 = plt.subplots(figsize=(8,6))

line4, = ax2.plot(x, object_tran[:, 0])
line5, = ax2.plot(x, object_tran[:,1])

figure3, ax3 = plt.subplots(figsize=(8,6))

# line6, = ax3.plot(x, object_rot[:, 0])
# line7, = ax3.plot(x, object_rot[:,1])
# line8, = ax3.plot(x, object_rot[:, 2])
# line9, = ax3.plot(x, object_rot[:,3])
# line10, = ax3.plot(x, object_rot[:, 4])
# line11, = ax3.plot(x, object_rot[:,5])
# line12, = ax3.plot(x, object_rot[:, 6])
# line13, = ax3.plot(x, object_rot[:,7])
line14, = ax3.plot(x, object_rot[:, 8])

print(np.linalg.norm(object_tran[2000] - object_tran[0], ord=2))
# torque = raibot[' actualTorque_0_0'].values[43000:47000]
# line5, = ax2.plot(x, torque)
# idx = 0
# while(True):
#     if(torque[idx+7000] == 0):
#         break
#     idx +=1
# print(idx)
```
